Функция для связки пар слов:

```python
def make_pairs(ind_words):
    for i in range(len(ind_words) - 1):
        yield (ind_words[i], ind_words[i + 1])
pair = make_pairs(ind_words)
```

```python
word_dict = {}
for word_1, word_2 in pair:
    if word_1 in word_dict.keys():
        word_dict[word_1].append(word_2)
    else:
        word_dict[word_1] = [word_2]
word_dict
```

### Работа с окнами в Марковских цепях

```python
import random

class Dictogram(dict):
    def __init__(self, iterable=None):
        # Инициализируем наше распределение как новый объект класса,
        # добавляем имеющиеся элементы
        super(Dictogram, self).__init__()
        self.types = 0  # число уникальных ключей в распределении
        self.tokens = 0  # общее количество всех слов в распределении
        if iterable:
            self.update(iterable)

    def update(self, iterable):
        # Обновляем распределение элементами из имеющегося
        # итерируемого набора данных
        for item in iterable:
            if item in self:
                self[item] += 1
                self.tokens += 1
            else:
                self[item] = 1
                self.types += 1
                self.tokens += 1

    def count(self, item):
        # Возвращаем значение счетчика элемента, или 0
        if item in self:
            return self[item]
        return 0

    def return_random_word(self):
        random_key = random.sample(self, 1)
        # Другой способ:
        # random.choice(histogram.keys())
        return random_key[0]

    def return_weighted_random_word(self):
        # Сгенерировать псевдослучайное число между 0 и (n-1),
        # где n - общее число слов
        random_int = random.randint(0, self.tokens-1)
        index = 0
        list_of_keys = self.keys()
        # вывести 'случайный индекс:', random_int
        for i in range(0, self.types):
            index += self[list(list_of_keys)[i]]
            # вывести индекс
            if(index > random_int):
                # вывести list_of_keys[i]
                return list(list_of_keys)[i]
            
def make_markov_model(data):
    markov_model = dict()

    for i in range(0, len(data)-1):
        if data[i] in markov_model:
            # Просто присоединяем к уже существующему распределению
            markov_model[data[i]].update([data[i+1]])
        else:
            markov_model[data[i]] = Dictogram([data[i+1]])
    return markov_model

def make_higher_order_markov_model(order, data):
    markov_model = dict()

    for i in range(0, len(data)-order):
        # Создаем окно
        window = tuple(data[i: i+order])
        # Добавляем в словарь
        if window in markov_model:
            # Присоединяем к уже существующему распределению
            markov_model[window].update([data[i+order]])
        else:
            markov_model[window] = Dictogram([data[i+order]])
    return markov_model
```

```python
from collections import deque
import re

def generate_random_start(model):
    # Чтобы сгенерировать любое начальное слово, раскомментируйте строку:
    # return random.choice(model.keys())

    # Чтобы сгенерировать "правильное" начальное слово, используйте код ниже:
    # Правильные начальные слова - это те, что являлись началом предложений в корпусе
    if 'END' in model:
        seed_word = 'END'
        while seed_word == 'END':
            seed_word = model['END'].return_weighted_random_word()
        return seed_word
    return random.choice(list(model.keys()))


def generate_random_sentence(length, markov_model):
    current_word = generate_random_start(markov_model)
    sentence = [current_word]
    for i in range(0, length):
        current_dictogram = markov_model[current_word]
        random_weighted_word = current_dictogram.return_weighted_random_word()
        current_word = random_weighted_word
        sentence.append(current_word)
    sentence[0] = sentence[0].capitalize()
    return ' '.join(sentence) + '.'
    return sentence
```

## Markov chain generation of article abstracts

### Работа с английскими статьями (успешно)

```python
import numpy as np
import pandas as pd
import markovify
```

```python
df = pd.read_json("arxivData.json")

#concatenating titles and abstract in new column
df["all_text"] = df["title"] + ". " + df["summary"]
df["all_text"] = df["all_text"].map(lambda x : x.replace("\n", " "))
df.head(5)
```

```python
STATE_SIZE = 2

#generating a model for all the text and one only for titles
text_model = markovify.Text( df["all_text"], state_size=STATE_SIZE)
title_model = markovify.Text( df["title"], state_size=STATE_SIZE)
```

```python
def findnth( str, char=" ", n=2):
    """
    Returns position of n-th occurence of pattern in a string
    """
    
    index_from_beg = 0
    while n >= 1:
        index = str.find( char)
        str = str[index+1:]
        index_from_beg += index + len(char)
        n -= 1
    return index_from_beg

sample_size = 7
successes = 0
f = open('result.txt','w')
while successes < sample_size:
    try: #some make_sentence calls raise a KeyError exception for misunderstood reasons
        #first generating a title
        print(successes+1)
        f.write(f'{successes+1}')
        f.write('\n')
        _title = title_model.make_sentence()
        print("Сгенерированый заголовок:", _title)
        
        f.write("Сгенерированый заголовок: ")
        f.write(_title)
        f.write('\n')
            
        _end_of_title = " ".join( _title.split()[-STATE_SIZE:])

        #generating abstract from the end of the tile
        _abstract = text_model.make_sentence_with_start( _end_of_title)
        print("Аннотация:", _abstract)
        #concatenating both
        index = findnth( _abstract, " ", 2)
        _abstract = _abstract[index:]
        _full_article_description = _title + " " + _abstract
        print("Итог")
        print( _full_article_description, end="\n\n")
        
        f.write(_full_article_description)
        f.write('\n\n')
            
        successes += 1

    except:
        pass
f.close()
```

```python
from nltk.stem import SnowballStemmer
def Stemmer(abstract):
    ind_words = abstract.split()
    abstract_stem = []
    for str in ind_words:
        snowball = SnowballStemmer(language="russian")
        abstract_stem.append(snowball.stem(str))
    abstract_stem = " ".join(abstract_stem)
    return abstract_stem
abstract_stem = Stemmer(abstract)
ag1_stem = Stemmer(ag1)
ag2_stem = Stemmer(ag2)
print(abstract_stem, "\n")
print(ag2_stem)
```
